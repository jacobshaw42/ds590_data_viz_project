```python
df = df.rename(columns = columns_renamed)
```

```python
df[df['NAME'] == 'San Juan Municipio, Puerto Rico']
```

We need to include the years from each dataset that these values were pulled from.

But let's take a look at the % variance for each county to see how far off they are from each other.

```python
#df = pd.DataFrame({'GROUP': [1, 2, 1, 2, 1], 'VALUE': [5, 2, 10, 20, 7]})
result = df.groupby('NAME')['total_pop'].agg(['max','min','mean'])
result['diff'] = (result['max']-result['min'])/result['mean']
print(result['diff'].info())

print('Max percentage difference: '.format(result['diff'].max()))
print('Min percentage difference: '.format(min(result['diff'])))
print('Mean percentage difference: '.format(np.mean(result['diff'])))
```
