### My alternate version

Here is my alternate dataset, which should be identical to Jacob's.

First, I put the dataframes into a list, and have a separate list of the corresponding years.

```python
df_list = [df_2021,df_2022,df_2019,
           df_2018,df_2015,df_2014,
           df_2016,df_2017,df_2013,
           df_2012,df_2010,df_2011]
```

```python
df_list_names = ['2021','2022','2019',
                 '2018','2015','2014',
                 '2016','2017','2013',
                 '2012','2010','2011']
```

Then I iterate over the dataframes to remove the first row, then add a new column which is the year of each respective dataframe.

```python
for i in range(len(df_list)):
    df_list[i] = df_list[i].iloc[1:]
    df_list[i]['Year'] = df_list_names[i]
    #print(i[j])
    
```

```python
df_list[0].head(1)
```

Once the dataframes are prepped, we combine them together row-wise.

```python
df_merged = pd.concat(df_list)
```

```python
df_merged.info()
```

```python
df_merged.head(2)
```

```python
columns_to_keep.append('Year')
```

```python
columns_to_keep[-1]
```

For the new dataframe, we'll remove all columns except the specified 14 columns.

```python
df_merged = df_merged[columns_to_keep]
df_merged = df_merged.rename(columns = columns_renamed)
df_merged.info()
```

```python
df_merged.sort_values(by = ['NAME', 'Year'], ascending =  True, inplace = True)
```

```python
#df_merged.to_excel('Final Project Dataset - Schmig.xlsx')
```

```python
num_cols = df_merged.columns[1:]
num_cols
```

Now that our data is aggregated, we need to remove values that aren't numeric.  In the dataset, there is a possibility for the value to be "N" instead of a number because the "entry in the estimate and margin of error columns indicates that data for this geographic area cannot be displayed because the number of sample cases is too small." (U.S. Census Bureau data notes)  As such, I will replace the "N" values by filling the value with the previous year's value.

```python
df_merged = df_merged.replace("N", method="ffill")
```

```python
num_cols = df_merged.columns[1:]
num_cols
```

```python
df_merged[num_cols] = df_merged[num_cols].astype(int)
```
